```python
import random
import matplotlib.pyplot as plt

# Initialize position
x, y = 0, 0
steps = 100  # Number of steps

# Store positions for plotting
positions = [(x, y)]

for _ in range(steps):
    # Generate random movement in x and y directions
    move_x = random.choice([-1, 1])  # Move left or right
    move_y = random.choice([-1, 1])  # Move up or down

    x += move_x
    y += move_y
    positions.append((x, y))

# Plot the path
x_values, y_values = zip(*positions)
plt.plot(x_values, y_values, marker='o')
plt.title('Random Movement in 2D Space')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid()
plt.axis('equal')
plt.show()
```
